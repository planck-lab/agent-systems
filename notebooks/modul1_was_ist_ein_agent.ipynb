## Die 5 Komponenten eines Agents

Nach Wooldridge & Jennings (1995) hat jeder Agent:
1. **Ziel** — Was will der Agent erreichen?
2. **Wahrnehmung** — Was nimmt er wahr? (Input)
3. **Reasoning** — Wie entscheidet er? (LLM)
4. **Aktion** — Was tut er? (Tools, Output)
5. **Gedächtnis** — Was merkt er sich?

```python
# Definiere die 5 Komponenten als Dictionary
agent_components = {
    'ziel': 'Was der Agent erreichen will',
    'wahrnehmung': 'Input (Text, Dateien, API-Responses)',
    'reasoning': 'Entscheidungslogik (LLM, Regeln)',
    'aktion': 'Output (Text, Tool-Aufrufe, Dateien)',
    'gedaechtnis': 'Kurzzeit (Kontext) + Langzeit (gespeichert)'
}

for k, v in agent_components.items():
    print(f'  {k.upper():15s} → {v}')
```
